```python
from os.path import join

english_yomi_paths = []

for sample in range(1,45):

    english_yomi_paths.append(join("..","jk",f"E{sample:02}.wav"))

print(english_yomi_paths)
```

```python

japanese_yomi_paths = []

for sample in range(1,45):

    japanese_yomi_paths.append(join("..","jk",f"J{sample:02}.wav"))

print(japanese_yomi_paths)
```

```python
from scipy.io.wavfile import read

english_yomi = []

for path in english_yomi_paths:

    eigo_sample,eigo_onsei = read(path)

    english_yomi.append(eigo_onsei)

print (english_yomi)





```

```python

japanese_yomi = []

for path in japanese_yomi_paths:

    nihongo_sample,nihongo_onsei = read(path)

    japanese_yomi.append(nihongo_onsei)

print (japanese_yomi)

```
